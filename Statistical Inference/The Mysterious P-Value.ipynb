Question:

A simple random sample of 1028 US adults show that 56% support clean energy. Does this provide convincing evidence that a majority of Americans supported clean energy at the 5% significance level?

Hypothesis:

$H_{0}$: Half of Americans support clean energy. $p_{0} = 0.5$

$H_{a}$: A majority of Americans support or oppose clean energy. $p_{0} \neq 0.5$

Let $\hat{p} = 0.56$ denote the sample proportion.

Assume the null hypothesis $H_{0}$ is true.

Since the sample is random, the observations are independent (Independence)

Success-failure condition:

$n * p_{0} = 1028 * 0.5 = 514 \geq 10$

$n * (1 - p_{0}) = 1028 * (1 - 0.5) = 514 \geq 10$

$\hat{p} \sim N(\mu_\hat{p} = p_{0}, SE_\hat{p} = \sqrt{\frac{p_{0} * (1 - p_{0})}{n}})$ (Central Limit Theorem)

$\sim N(\mu_\hat{p} = 0.5, SE_\hat{p} = \sqrt{\frac{0.5 * (1 - 0.5)}{1028}})$

$\sim N(\mu_\hat{p} = 0.5, SE_\hat{p} = 0.0156)$

$\sim \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}(\frac{x - \mu}{\sigma})^{2}}$

$\sim \frac{1}{0.0156\sqrt{2\pi}} e^{-\frac{1}{2}(\frac{x - 0.5}{0.0156})^{2}}$

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_density(x, mean, sd):
    return (1/(sd*np.sqrt(2*np.pi)))*np.exp((-1/2)*((x-mean)/sd)**2)

x = np.linspace(0, 1)
y = normal_density(x, 0.5, 0.0156)
plt.figure(figsize=(14,7))
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title ('Normal Distribution of Sampling Proportion under Null Assumption P_0 = 0.5')
plt.plot(x, y)
```

$z = \frac{\hat{p} - \mu}{\sigma} = \frac{\hat{p} - \mu_\hat{p}}{SE_\hat{p}} = \frac{0.56 - 0.5}{0.0156} = 3.85$

$p = F(z \leq -3.85) + F(3.85 \leq z)$ (lower tail area + upper tail area)

$= \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{-3.85} e^{-\frac{z^{2}}{2}} dz + \frac{1}{\sqrt{2\pi}} \int_{3.85}^{\infty} e^{-\frac{z^{2}}{2}} dz $

$= F(-3.85) + (1 - F(3.85))$

```python
import scipy.stats as st
print("P-Value = {0}".format(st.norm.cdf(-3.85) + (1 - st.norm.cdf(3.85))))
```

Since P-Value = 0.0001181178248378968 < 0.05, it is very unlikely to observe the sample $\hat{p} \geq 0.56$ (majority support) or $\hat{p} \leq 0.44$ (majority oppose) if the null hypothesis $H_{0}$ is true, $H_{0}: p_{0} = 0.5$ is rejected.
